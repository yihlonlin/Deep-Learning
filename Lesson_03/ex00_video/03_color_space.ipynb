```python
#---------------------------------------
# required packages
#---------------------------------------
import cv2
import numpy as np
from matplotlib import pyplot as plt
#===========================================================
# Color space
#===========================================================
#---------------------------------------
# Y Cb Cr
#---------------------------------------
#fn_in="lena.tiff"
#fn_in="baboon.bmp"
#fn_in="landscape.jpg"
fn_in="image02/rainbow_draw.png"

img_in = cv2.imread(fn_in)
#cv2.imshow('Input Image', img_in)
#cv2.waitKey(0)
plt.imshow(img_in[...,::-1])  
plt.show()
```

```python
 #BGR to YCbCr
YCC = cv2.cvtColor(img_in,cv2.COLOR_BGR2YCR_CB)
type(YCC)
YCC.shape
Y = YCC[:,:,0]
Y.shape
Y.dtype
Cr = YCC[:,:,1]
Cr.shape
Cb = YCC[:,:,2]
Cb.shape

#cv2.imshow('Y',Y)
#cv2.imshow('Cr',Cr)
#cv2.imshow('Cb',Cb)
#cv2.waitKey(0)

plt.imshow(Y,cmap='Greys_r')  
plt.show()
plt.imshow(Cb,cmap='Greys_r')
plt.show()
plt.imshow(Cr,cmap='Greys_r')  
plt.show()

cv2.imwrite('Y.bmp',Y);
cv2.imwrite('Cr.bmp',Cr);
cv2.imwrite('Cb.bmp',Cb);

#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
#---------------------------------------
# HSV
#---------------------------------------
#---------------------------------------
# required packages
#---------------------------------------
import cv2
import numpy as np
from matplotlib import pyplot as plt
#fn_in="lena.tiff"
#fn_in="baboon.bmp"
#fn_in="landscape.jpg"
fn_in="image02/rainbow_draw.png"

img_in = cv2.imread(fn_in)
#cv2.imshow('Input Image', img_in)
#cv2.waitKey(0)
plt.imshow(img_in[...,::-1])  
plt.show()
#img_in = cv2.imread(fn_in)
#cv2.imshow('Input Image', img_in)
#cv2.waitKey(0)
```

```python
#BGR to HSV
HSV = cv2.cvtColor(img_in,cv2.COLOR_BGR2HSV)
type(HSV)
HSV.shape

H = HSV[:,:,0]
H.shape
H.dtype 
S = HSV[:,:,1]
S.shape
V = HSV[:,:,2]
V.shape

#cv2.imshow('H',H)
#cv2.imshow('S',S)
#cv2.imshow('V',V)
#cv2.waitKey(0)

plt.imshow(H,cmap='Greys_r')  
plt.show()
plt.imshow(S,cmap='Greys_r')
plt.show()
plt.imshow(V,cmap='Greys_r')  
plt.show()

cv2.imwrite('H.bmp',H);
cv2.imwrite('S.bmp',S);
cv2.imwrite('V.bmp',V);
#cv2.waitKey(0)
#cv2.destroyAllWindows()
#===========================================================
```
